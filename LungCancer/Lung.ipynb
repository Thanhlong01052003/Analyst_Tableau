```python
import pandas as pd
```

```python
df = pd.read_csv('cancer patient data sets.csv')
df.drop(columns=['Patient Id'], inplace=True)
df.drop(columns=['index'], inplace=True)
```

```python
col = df.columns.tolist()
col
```

```python
Age = [df['Age'].tolist()]
Gender = [df['Gender'].tolist()]
Air_Pollution = [df['Air Pollution'].tolist()]
Alcohol_use = [df['Alcohol use'].tolist()]
Dust_Allergy = [df['Dust Allergy'].tolist()]
Occupational_Hazards = [df['OccuPational Hazards'].tolist()]
Genetic_Risk = [df['Genetic Risk'].tolist()]
Chronic_Lung_Disease = [df['chronic Lung Disease'].tolist()]
Balanced_Diet = [df['Balanced Diet'].tolist()]
Obesity = [df['Obesity'].tolist()]
Smoking = [df['Smoking'].tolist()]
Passive_Smoker = [df['Passive Smoker'].tolist()]
Chest_Pain = [df['Chest Pain'].tolist()]
Coughing_of_Blood = [df['Coughing of Blood'].tolist()]
Fatigue = [df['Fatigue'].tolist()]
Weight_Loss = [df['Weight Loss'].tolist()]
Shortness_of_Breath = [df['Shortness of Breath'].tolist()]
Wheezing = [df['Wheezing'].tolist()]
Swallowing_Difficulty = [df['Swallowing Difficulty'].tolist()]
Clubbing_of_Finger_Nails = [df['Clubbing of Finger Nails'].tolist()]
Frequent_Cold = [df['Frequent Cold'].tolist()]
Dry_Cough = [df['Dry Cough'].tolist()]
Snoring = [df['Snoring'].tolist()]
Level = [df['Level'].tolist()]
```
